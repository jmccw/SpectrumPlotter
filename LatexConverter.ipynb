```python
from tabulate import tabulate
import regex as re
import tkinter
from tkinter import filedialog

#tkinter.Tk().withdraw() # prevents an empty tkinter window from appearing
path = filedialog.askopenfilename()

# Read data from a file - TAB DELIMINATED '*.txt' FILES
with open(path, 'r') as file:
    data = [line.strip().split('\t') for line in file]
    
# Separate headers and data
headers = data[0]
data = data[1:]

# Add LaTeX bold formatting to the headers
headers = [r"""\textbf{""" + header + """}""" for header in headers]

# Combine headers and data
data.insert(0, headers)

# Generate LaTeX code for the table
latex_table = tabulate(data, headers="firstrow", tablefmt="latex_booktabs")

# Print the LaTeX code
# Wrap the LaTeX table in a minimal LaTeX document
latex_print = r"""
\begin{table}[htbp]
  \centering
  \label{tab:sample-table}
  """ + latex_table + r"""
  \caption{Sample Data}
\end{table}
"""

# Fix formatting for pasting code
latex_print = re.sub(r'textbackslash{}', r'', latex_print)
latex_print = re.sub(r'textbf\\', r'textbf', latex_print)
latex_print = re.sub(r'\\}', r'}', latex_print)

# Print the complete LaTeX document
print(latex_print)
print("\n\nMake sure to include: \n\\usepackage{array}\n\\usepackage{bookt\
abs}\n\\usepackage{caption}")
```
